# Raw Commodities Loader
This notebook is for downloading commodity data.
There are two big limitations:
- 100 monthly calls
- Maximum granularity: Day by Day


```python
import json
import requests
from datetime import date
from datetime import timedelta
import datetime as dt
import pandas as pd
```

```python
api_key = 'TU_LLAVE'
api_url_base = 'https://commodities-api.com/api/timeseries'
```

```python
all_symbols = [
    'ALU' #alumunium
    , 'COFFEE' #arabica coffee
    , 'BRENTOIL' #brent oil
    , 'CANO' # canola
    , 'COCOA' # cocoa
    , 'XCU' #COPPER
    , 'CORN' #corn
    , 'COTTON' #Cotton
    , 'ETHANOL' #ETHANOL
    , 'XAU' # GOLD
    , 'IRD' #Iridium
    , 'NG' #natural gas
    , 'NI' #nickel
    , 'OAT' #oat
    , 'RICE' #rice
    , 'ROBUSTA' # robusta coffee
    , 'RUBBER' # 
    , 'SOYBEAN'
    , 'SUGAR'
    , 'TIN' # lata
    , 'WHEAT' #harina
    , 'ZNC' # zinc

]
```

```python
c = 0


loop_start_time = dt.datetime.now()

for com in all_symbols:
    # Set API Parameters
    par = {'access_key': api_key, 'start_date': '2022-01-01'
            , 'end_date': '2022-12-31', 'base': 'USD', 'symbols':com}

    # Call from API
    req = requests.get(api_url_base, params= par)

    # Check if valid call
    if not req.ok:
        print(f"Error in the request for {com}\nCheck req var status.")
        break
    
    # API response to dictionary
    res = json.loads(req.text)

    if not res['data']['success'] :
        print(f"Not successful call for {com}\nCheck res var status.")
        print(f"Error: {res['data']['error']['info']}")
        break

    rates = res['data']['rates']
    unit = 'per_unit'
    if res['data'].get('unit') is not None:
        unit = res['data']['unit'].replace(' ','_')
    

    df = pd.DataFrame.from_dict(rates).transpose()
    #df.rename(columns=['d_vintage', 'XAU','USD'], inplace=True)
    df['DATE'] = df.index
    
    path = f"data/commodities/raw_{com}_{unit}"

    df.to_csv(path)

    c+=1

loop_end_time = dt.datetime.now()

time = loop_end_time - loop_start_time
print('DONE'
    , '\nAPI Calls: ',c
    , '\nDownloaded Rows: ', (c*365)
    , '\nDURATION:', time.days, 'DAYS'
    , time.seconds//3600, 'HOURS', (time.seconds//60)%60, 'MINUTES' 
    , time.seconds%60 , 'SECONDS.')
print(c)
```
