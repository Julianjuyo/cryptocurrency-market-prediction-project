## Commodities Null Prices

```python
commodities = [
 'ALU',
 'CANO',
 'WHEAT',
 'XCU',
 'BRENTOIL',
 'COFFEE',
 'SOYBEAN',
 'ROBUSTA',
 'NG',
 'NI',
 'RICE',
 'OAT',
 'XAU',
 'COCOA',
 'IRD',
 'SUGAR',
 'ETHANOL',
 'CORN',
 'RUBBER',
 'COTTON',
 'TIN',
 'ZNC']
```

```python
for element in commodities:
    df[element] = df[element].fillna(method='ffill')
```

```python
# Set to none to see all
pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 30)
```

```python
c = df.datetime.count()

tddf = df.describe().transpose()

tddf['null_flag'] = tddf['count'] < c
tddf['nnull_share'] = tddf['count']/c

tddf.sort_values(by='nnull_share').head(30)

to_drop = tddf[tddf['nnull_share'] < 0.99].index.tolist()
to_drop
```

```python
df.drop(columns=to_drop, inplace=True)
```
