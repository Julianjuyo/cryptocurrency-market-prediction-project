```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('final_dataset22.csv')
df
```

```python
df.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
```

```python
transformers = [
    ('minmax', MinMaxScaler(), [col for col in df.columns]),
]

# Define a column transformer to apply the normalization transformers to the columns
ct = ColumnTransformer(transformers=transformers)
```

```python
# Fit and transform the dataframe using the column transformer
df_normalized = ct.fit_transform(df)

# Convert the transformed data array back to a dataframe with the original column names
df_normalized = pd.DataFrame(df_normalized, columns=df.columns)

print(df_normalized)
```

```python
df = df_normalized
```
